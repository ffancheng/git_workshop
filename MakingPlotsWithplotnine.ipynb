 import matplotlib.pyplot as plt


```
%matplotlib inline
survs_df = pd.read_csv('surveys.csv').dropna()
survs_df.describe()
```

```
ggplot(survs_df, aes(x='weight', y='hindfoot_length')) + geom_point()
```

```
ggplot(survs_df, aes(x='weight', y='hindfoot_length', size = 'year')) + geom_point()
```

```
ggplot(survs_df, aes(x='weight', y='hindfoot_length', size = 'year', color = "species_id")) + geom_point()
```

```
ggplot(survs_df, aes(x = 'year')) + \
       geom_bar(stat = 'count')
```

```
ggplot(survs_df, aes('year', fill = 'species_id')) + \
       geom_bar(stat = 'count')
```
